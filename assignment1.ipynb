```python
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display
```

```python
ylabel='Amplitude'
xlabel='Time (Seconds)'
file='avm.wav'
```

```python
sig , sr = librosa.load(file,mono=True,sr=44100)
time=np.arange(0,sig.shape[0])/sr
excursion=max(abs(sig))
sigNew = sig / excursion
plt.plot(time,sigNew)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
```
